```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = "median") # replace them with mean values

# create a copy data frame without the text coloumn ocean proximity
housing_num = housing.drop("ocean_proximity", axis = 1)

imputer.fit(housing_num) # learn and calculate the median on each meadian coloumn(fit method)

print(imputer.statistics_) # stock the medians

X = imputer.transform(housing_num) # replaces NA values with median

print(housing_num.columns) # shows the columns

np.set_printoptions(suppress = True)
print("X: {}".format(X))

# change back to array into a DF
housing_tr = pd.DataFrame(X, columns = housing_num.columns,
                             index = housing_num.index)

housing_tr # show the table
```

```python
# columns 3 to 6 from array X
print('Rooms in each district (column 3):', X[:, 3])
print('Bedrooms in each district (column 4):', X[:, 4])
print('Population in each district (column 5):', X[:, 5])
print('HouseHolds in each district (column 6):', X[:, 6])
```

```python
# creating a transformer class
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin): # add new attributes witch they are calculated combined with existing attributes
    def __init__(self, add_bedrooms_per_room = True): # default add bedrooms per room boolean value (function constructor)
        self.add_bedrooms_per_room = add_bedrooms_per_room
    
    def fit(self, X, y = None):
        return self # that means that nothing happends with the object
    
    def transform(self, X):
        rooms_per_households = X[:, rooms_ix] / X[:, households_ix] # create first column
        population_per_household = X[:, population_ix] / X[:, households_ix] # second column
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]

            # concatenate  new columns to the right side of matrix X, using numpy
            return np.c_[X, rooms_per_households, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_households, population_per_household]

# test how the class performs:

attr_adder = CombinedAttributesAdder()

housing_extra_attribs = attr_adder.transform(housing.values)

print(housing_extra_attribs[:, -4::]) # printing the last 4 columns

attr_adder = CombinedAttributesAdder(False)

print('\n')
housing_extra_attribs = attr_adder.transform(housing.values)

print(housing_extra_attribs[:, -4::]) # printing the last 4 columns

```

```python
# pipeline

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# create an object with the class pipeline (list of tuples)
num_pipeline = Pipeline([("imputer", SimpleImputer(strategy = "median")), # fill the missing values with the median 
                         ("attribs_adder", CombinedAttributesAdder()), # add attributes with the transformer class
                         ("std_scaler", StandardScaler())]) # scale all of the numbers

housing_num_tr = num_pipeline.fit_transform(housing_num) # train (fit to find the median) and modify the values with median the list of tuples

# printing first 5 rows before transformed
print(housing_num.head(5), '\n')

# housing transformed and modified
print(housing_num_tr[:6, :].round(2))
```

```python
# transforming numerical and categorical columns in a single pipeline

from sklearn.compose import ColumnTransformer # to work with strings too (pipeline works only with numerical values)

# first, separate numerical from text attributes
num_attribs = list(housing_num) # creating a list with the names of the columns from DataFrame
cat_attribs = ["ocean_proximity"] # list with a single column name 5 categories

# create an object
full_pipeline = ColumnTransformer([("num", num_pipeline, num_attribs), # applies numerical pipeline on the list 
                                   ("cat", OneHotEncoder(), cat_attribs)]) # transforms ocean_proximity to numerical (0 and 1)
# prepare the training model (housing) for final training
housing_prepared = full_pipeline.fit_transform(housing)

# print firsts 5 rows and 6 columns from the full pipeline
print(housing_prepared[:6, :])
```

```python
# TRAINING MODELS

# METHOD 1- SIMPLE TRAINING(the worse method)

# train and evaluate the model using linear regression
from sklearn.linear_model import LinearRegression

# object
lin_reg = LinearRegression()

# train the model
lin_reg.fit(housing_prepared, housing_labels) # matrix and array
```

```python
# predict median for 5 rows (districts)

some_data = housing.head() # select first 5 rows
some_labels = housing_labels.head() # select first 5 prices

# prepare the data before trying the model
some_data_prepared = full_pipeline.transform(some_data) 

print("Predicted prices", lin_reg.predict(some_data_prepared).round(0)) # predicted prices
print()
print("Actual prices: ", list(some_labels))
print()
print("Absolute error (predicted - actual)", lin_reg.predict(some_data_prepared).round(0) - list(some_labels))
```

```python
# calculate the error RMSE (Root Mean Squared Error)

from sklearn.metrics import mean_squared_error

housing_predictions = lin_reg.predict(housing_prepared)

print(housing_predictions) # print predictions
```

```python
lin_mse = mean_squared_error(housing_labels, housing_predictions) # just calculate the values in the square root

# and calculate with the square root
lin_rmse = np.sqrt(lin_mse)

print(lin_rmse) # 68628 - too large (makes underfitting) NOT A GOOD MODEL LINEAR REGRESSION
```

```python
# using a more powerful model for better results - DecissionTreeRegressor

from sklearn.tree import DecisionTreeRegressor

# object to be trained
tree_reg = DecisionTreeRegressor(random_state = 42) # makes the random be all the time the same

tree_reg.fit(housing_prepared, housing_labels) # to make the training

# evaluate the trained model on the entire prepared training set
housing_predictions = tree_reg.predict(housing_prepared)

# find the RMSE
tree_mse = mean_squared_error(housing_labels, housing_predictions)

tree_rmse = np.sqrt(tree_mse)

print(tree_rmse) # is not better, is worse, it learning it too perfectly
```

```python
# METHOD 2 - cross validation (a little bit better method)

from sklearn.model_selection import cross_val_score # fit and predict

scores = cross_val_score(tree_reg, housing_prepared, housing_labels, scoring = "neg_mean_squared_error", cv = 10)

print(scores) # negatives scores 
```

```python
# compute RMSE
tree_rmse_scores = np.sqrt(-scores)

print(tree_rmse_scores.round(0))
```

```python
# define a function to display errors and calculate the mean of the erros and the std. deviation
def display_scores(scores):
    print("Scores: ", scores.round(0))
    print("Mean of scores: ", scores.mean().round(0))
    print("Scores' std deviation:", scores.std().round(0))

display_scores(tree_rmse_scores) # positive values
```

```python
# compute cross validation for linear regression

lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels, scoring = "neg_mean_squared_error", cv = 10)

lin_rmse_scores = np.sqrt(-lin_scores)

display_scores(lin_rmse_scores.round(0))
```

```python
# METHOD 3 - RANDOM FOREST REGRESSOR THE BEST METHOD

# calculate the error without cross validation
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(random_state = 42) # to make the same random selection

forest_reg.fit(housing_prepared, housing_labels)

housing_predictions = forest_reg.predict(housing_prepared) # making predictions on the same matrix

forest_mse = mean_squared_error(housing_labels, housing_predictions)

forest_rmse = np.sqrt(forest_mse)

print("Forest rmse: ", forest_rmse.round(0)) # the difference is too big between method 1 & method 2
```

```python
# calculate error with cross validation BEST SOLUTION

forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels, scoring = "neg_mean_squared_error", cv = 10)

forest_rmse_scores = np.sqrt(-forest_scores)

display_scores(forest_rmse_scores)
```

```python
# find the best values for hyperparameters with GridSearchCV for a small number of hyperparameters / RandomizedSearchCV for a large number of hyperparameters

# using GridSearchCV
# if you have no idea of hyperparameter, try consecutive power of 10, or smaller values
from sklearn.model_selection import GridSearchCV

# making the list
param_grid = [{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]}, 
              {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]}]
```

```python
# making 90 rounds of training {3 x 4} + {2 x 3} x 5
forest_reg = RandomForestRegressor(random_state = 42, n_jobs = -1) # n_jobs to put all the 8 cores to join the training

# create the grid search object
grid_search = GridSearchCV(forest_reg, param_grid, cv = 5, scoring='neg_mean_squared_error', return_train_score = True)

%time grid_search.fit(housing_prepared, housing_labels) # shows the duration
```

```python
print(grid_search.best_estimator_) # shows the model name

print(grid_search.best_params_)  # shows only the best combination
```

```python
# calculate the error
forest_reg = RandomForestRegressor(max_features=8, n_estimators=30, n_jobs=-1, random_state=42)
forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels, scoring = "neg_mean_squared_error", cv = 5)

forest_rmse_scores = np.sqrt(-forest_scores)

display_scores(forest_rmse_scores) # good solution
```

```python
# using RandomizedSearchCV

from sklearn.model_selection import RandomizedSearchCV

param_rnd_src = [{'bootstrap': [True, False], 'n_estimators': list(range(10, 34)), 'max_features': list(range(6, 10))}]

rand_src = RandomizedSearchCV(forest_reg, param_rnd_src, n_iter = 50, cv = 5, scoring = "neg_mean_squared_error", return_train_score = True, n_jobs = -1, random_state = 42)

%time rand_src.fit(housing_prepared, housing_labels)
```

```python
# print the best estimators and params
print(rand_src.best_estimator_)

print(rand_src.best_params_, '\n')
```
